```python
import gzip
import shutil
import os

def amaliy_ish_bajarish():
    # 1. Asl matnli ma'lumotni shakllantirish
    matn = b"Bu ma'lumotlarni siqish orqali saqlash usuli. " * 100

    # 2. Ma'lumotni asl variantda faylga yozish
    asl_fayl = 'data.txt'
    with open(asl_fayl, 'wb') as f:
        f.write(matn)

    # 3. Gzip yordamida yo'qotishsiz siqish (Lossless Compression)
    siqilgan_fayl = 'data.txt.gz'
    with open(asl_fayl, 'rb') as f_in:
        with gzip.open(siqilgan_fayl, 'wb') as f_out:
            shutil.copyfileobj(f_in, f_out)

    print("Siqish jarayoni muvaffaqiyatli yakunlandi.")

    # 4. Siqilgan .gz faylini ochish va o'qish
    with gzip.open(siqilgan_fayl, 'rb') as f_ochilgan:
        tiklangan_matn = f_ochilgan.read()

    # 5. Statistik ma'lumotlarni hisoblash
    asl_hajm = os.path.getsize(asl_fayl)
    siqilgan_hajm = os.path.getsize(siqilgan_fayl)
    samaradorlik = (1 - (siqilgan_hajm / asl_hajm)) * 100

    # 6. Natijalarni terminalga chiqarish
    print("\n========= TAHLIL NATIJALARI =========")
    print(f"Asl fayl hajmi: {asl_hajm} bayt")
    print(f"Siqilgan fayl hajmi: {siqilgan_hajm} bayt")
    print(f"Xotira tejalishi: {samaradorlik:.2f}%")

    if matn == tiklangan_matn:
        print("Butunlik tekshiruvi: Ma'lumotlar yo'qotishsiz to'liq tiklandi.")
    else:
        print("Butunlik tekshiruvi: Xatolik aniqlandi.")

if __name__ == "__main__":
    amaliy_ish_bajarish()
```
